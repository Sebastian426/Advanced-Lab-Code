```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.optimize import curve_fit
from scipy.special import voigt_profile
```

```python
#helper functions

def voigt_FWHM(gamma, sigma):
    Gamma_L = 2 * gamma
    Gamma_G = 2 * np.sqrt(2 * np.log(2)) * sigma
    Gamma_V = 0.5346 * Gamma_L + np.sqrt(0.2166 * Gamma_L**2 + Gamma_G**2)
    return Gamma_V
```

```python
###### data #######

# 0: 77
# 1: 90
# 2: 120
# 3: 273

sigma1 = np.array([5.38169902e-01, 4.80964229e-01, 5.10662533e-01, 4.35955673e-01])
sigma2 = np.array([4.43602405e+02 , 1.58846697e+02, 8.22028021e+01, 4.62686858e-01])
gamma = np.array([2.12368914e-01, 3.15666255e-01, 2.87093817e-01, 9.13514985e-02])
temperatures = np.array([77, 90, 120, 273])
```

```python
FWHM_R1 = voigt_FWHM(gamma, sigma1)
FWHM_R2 = voigt_FWHM(gamma, sigma2)
```

```python
plt.figure(1)
plt.scatter(temperatures, FWHM_R1, color = "red")
plt.xlabel("Temperature (K)")
plt.ylabel("FWHM")
plt.title("R1")
plt.show()


plt.figure(1)
plt.scatter(temperatures, FWHM_R2, color = "blue")
plt.xlabel("Temperature (K)")
plt.ylabel("FWHM")
plt.title("R2")
plt.show()
```
