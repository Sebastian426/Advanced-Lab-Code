```python
import Ocean_Opt_Algorithm as OOA
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import importlib
from matplotlib.patches import Ellipse

def covariance_ellipse(ax, x, y, n_std=1, **kwargs):

    x = np.asarray(x)
    y = np.asarray(y)

    cov = np.cov(x, y)

    # Eigen-decomposition
    vals, vecs = np.linalg.eigh(cov)

    # Sort largest → smallest
    order = vals.argsort()[::-1]
    vals = vals[order]
    vecs = vecs[:, order]

    # Ellipse angle
    theta = np.degrees(np.arctan2(vecs[1,0], vecs[0,0]))

    # Width and height
    width  = 2 * n_std * np.sqrt(vals[0])
    height = 2 * n_std * np.sqrt(vals[1])

    ellipse = Ellipse(
        (np.mean(x), np.mean(y)),
        width,
        height,
        angle=theta,
        fill=False,
        **kwargs
    )

    ax.add_patch(ellipse)

importlib.reload(OOA)
```

# Calculations

```python
#Linewidth function
def voigt_FWHM(sigma, gamma):
    gamma_L = 2*gamma
    gamma_G = 2*np.sqrt(2*np.log(2))*sigma
    return 0.5346*gamma_L + np.sqrt(0.2166*gamma_L**2+gamma_G**2)

from scipy.integrate import quad
Td_fixed = 760
def debye_integrand(x):
    return x**3 / (np.exp(x) - 1)

def debye_integral(T, Td):

    if T <= 0:
        return 0  # avoid divide by zero

    upper = min(Td/T, 50)

    result, _ = quad(
        lambda x: x**3/(np.exp(x)-1),
        0,
        upper
    )

    return result

def ruby_shift_model(T, R0, alpha, Td):
    integral_vals = np.array([debye_integral(t, Td) for t in T])
    return R0 + alpha * (T / Td)**4 * integral_vals

def ruby_shift_model_fixed(T, R0, alpha):
    integral_vals = np.array([debye_integral(t, Td_fixed) for t in T])
    return R0 + alpha * (T / Td_fixed)**4 * integral_vals
```

```python
temperatures = np.array(sorted(k for k in results.keys() if isinstance(k, (int, float))))

# Observed peak positions
R1_values = np.array([results[T]["R1_obs"] for T in temperatures])
R1_errs   = np.array([results[T]["R1_err"] for T in temperatures])

R2_values = np.array([results[T]["R2_obs"] for T in temperatures])
R2_errs   = np.array([results[T]["R2_err"] for T in temperatures])


# Voigt parameters
R1_sigma = np.array([results[T]["parameters"][3] for T in temperatures])
R1_sigma_errs = np.array([np.sqrt(results[T]["covariance"][3][3]) for T in temperatures])

R1_gammas = np.array([results[T]["parameters"][2] for T in temperatures])
R1_gamma_errs = np.array([np.sqrt(results[T]["covariance"][2][2]) for T in temperatures])


R2_sigma = np.array([results[T]["parameters"][7] for T in temperatures])
R2_sigma_errs = np.array([np.sqrt(results[T]["covariance"][7][7]) for T in temperatures])

R2_gammas = np.array([results[T]["parameters"][6] for T in temperatures])
R2_gamma_errs = np.array([np.sqrt(results[T]["covariance"][6][6]) for T in temperatures])


# Line shifts
R1_shift = R1_values[-1] - R1_values
R1_shift_err = np.sqrt(R1_errs[-1]**2 + R1_errs**2)

R2_shift = R2_values[-1] - R2_values
R2_shift_err = np.sqrt(R2_errs[-1]**2 + R2_errs**2)


# Linewidths
R1_linewidth = voigt_FWHM(R1_sigma, R1_gammas)
R1_linewidth_err = np.sqrt(
    (voigt_FWHM(R1_sigma + R1_sigma_errs, R1_gammas) - R1_linewidth)**2 +
    (voigt_FWHM(R1_sigma, R1_gammas + R1_gamma_errs) - R1_linewidth)**2
)

R2_linewidth = voigt_FWHM(R2_sigma, R2_gammas)
R2_linewidth_err = np.sqrt(
    (voigt_FWHM(R2_sigma + R2_sigma_errs, R2_gammas) - R2_linewidth)**2 +
    (voigt_FWHM(R2_sigma, R2_gammas + R2_gamma_errs) - R2_linewidth)**2
)
```

```python
# Results needed:
# R1, ∆R1, linewidths, R2, ∆R2, linewidths, chi2 theoretical, chi2 empirical
# 77, 130, 217, 298


def empirical_peakshift(T, a1, a2, a3, offset):
    return a1 * T + a2 * T ** 2 + a3 * T ** 3 + offset

def reduced_chi_squared_calc(model, model_params, xdata, ydata, yerror):
    dof = len(xdata) - len(model_params)
    theoretical = model(xdata, *model_params)
    chi2 = np.sum((theoretical - ydata) ** 2 / (yerror ** 2))
    rchi2 = chi2 / dof
    return chi2, rchi2

p0_emp_R1 = [0.001, 0.0001, 0.000001, np.min(R1_values)]

pop_emp_R1, cov_emp_R1 = curve_fit(empirical_peakshift,
                                   temperatures,
                                   R1_values,
                                   sigma=R1_errs,
                                   absolute_sigma=True,
                                   p0=p0_emp_R1,
                                   maxfev=10000)

plt.figure(1)
plt.errorbar(temperatures, R1_values, yerr=R1_errs,
             color="black", marker="o", markersize=5, markerfacecolor="none", capsize=3, linestyle="none")
plt.plot(temperatures, empirical_peakshift(temperatures, *pop_emp_R1),
         linestyle="--", color="grey")

plt.xlabel("Temperature (K)")
plt.ylabel("Peak Position")
plt.show()
p0_emp_R2 = [0.001, 0.0001, 0.000001, np.min(R2_values)]

pop_emp_R2, cov_emp_R2 = curve_fit(empirical_peakshift,
                                   temperatures,
                                   R2_values,
                                   sigma=R2_errs,
                                   absolute_sigma=True,
                                   p0=p0_emp_R2,
                                   maxfev=10000)

plt.figure(1)
plt.errorbar(temperatures, R2_values, yerr=R2_errs,
             color="black", marker="o", markersize=5, markerfacecolor="none", capsize=3, linestyle="none")
plt.plot(temperatures, empirical_peakshift(temperatures, *pop_emp_R2),
         linestyle="--", color="grey")

plt.xlabel("Temperature (K)")
plt.ylabel("Peak Position")
plt.show()

chi2_R1_Theoretical, rchi2_R1_Theoretical = reduced_chi_squared_calc(ruby_shift_model_fixed, pop_R1_pos, temperatures,
                                                                     R1_values, R1_errs)
chi2_R1_Emp, rchi2_R1_Emp = reduced_chi_squared_calc(empirical_peakshift, pop_emp_R1, temperatures, R1_values, R1_errs)

chi2_R2_Theoretical, rchi2_R2_Theoretical = reduced_chi_squared_calc(ruby_shift_model_fixed, pop_R2_pos, temperatures,
                                                                     R2_values, R2_errs)
chi2_R2_Emp, rchi2_R2_Emp = reduced_chi_squared_calc(empirical_peakshift, pop_emp_R2, temperatures, R2_values, R2_errs)
print(f" ----- Chi2 -----")
print(f"R1 Theoretical: {rchi2_R1_Theoretical:.4f}")
print(f"R1 Emp: {rchi2_R1_Emp:.4f}")
print(f"R2 Theoretical: {rchi2_R2_Theoretical:.4f}")
print(f"R2 Emp: {rchi2_R2_Emp:.4f}")
print(temperatures)

print(f"----- Results Table -----")
print(f"R1 77: {R1_values[0]:.4f}")
print(f"R1 130: {R1_values[2]:.4f}")
print(f"R1 217: {R1_values[7]:.4f}")
print(f"R1 298: {R1_values[10]:.4f}")
print(f" ----- ------ ------")
print(f"R2 77: {R2_values[0]:.4f}")
print(f"R2 130: {R2_values[2]:.4f}")
print(f"R2 217: {R2_values[7]:.4f}")
print(f"R2 298: {R2_values[10]:.4f}")
```
