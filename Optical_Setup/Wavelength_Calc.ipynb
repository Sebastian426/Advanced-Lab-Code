```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def monochromator_calibration(x):
    A = 1.0095
    alpha_A = 0.0001

    B = 9091.0570
    alpha_b = 0.0289

    index = A*x + B

    uncertainty = np.abs((A+alpha_A)*x + (B+alpha_b)-index)

    print(index)
    return index, uncertainty
```

```python
wavelength_blue = 405

wavelength_red = 692.7

wavelength_red_laser = 632.8
```

```python
monochromator_calibration(wavelength_red_laser)
```

```python
monochromator_calibration(wavelength_blue)
```

```python
monochromator_calibration(wavelength_red)
```
